**Desafio I  - PCA**

```
df.grau.unique()
```

```
X = df[['turno', 'mensalidade', 'bolsa_integral_cotas',
       'bolsa_integral_ampla', 'bolsa_parcial_cotas', 'bolsa_parcial_ampla',
       'nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla',
       'nota_parcial_cotas']]
y = df['grau']
```

### Fase 1 - Padronização dos dados



```
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

y_std = LabelEncoder().fit_transform(y)

X['turno'] = LabelEncoder().fit_transform(X['turno'])
#print(X['turno'].head())
Xtmp = SimpleImputer(strategy='constant',fill_value= 0).fit_transform(X)
X_std = StandardScaler().fit_transform(Xtmp)
```

### Fase 2 - Geração da matriz de covariância

```
import numpy as np

mean_vec = np.mean(X_std, axis=0)
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
```

```
# cada coluna dessa é covariância de cada variável por ocorrência
print('Matriz de covariância: \n%s' %np.cov(X_std.T))
```

### Fase 3 - Geração de autovetores e autovalores

```
#observando pelo método da covariância
cov_mat = np.cov(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Autovetores \n%s' %eig_vecs)
print('\nAutovalores \n%s' %eig_vals)
```

```
#matriz de correlação
#observando por um método de correlação ... perceba que os resultados são os mesmos.
cor_mat1 = np.corrcoef(X_std.T)

eig_vals, eig_vecs = np.linalg.eig(cor_mat1)
print('Autovetores \n%s' %eig_vecs)
print('\nAutovalores \n%s' %eig_vals)
```

```
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Está na ordem descrescente!')
```

### Fase 4 - Computando os PCAs


```
from sklearn.decomposition import PCA
X = X.fillna(0)

#PCA sem dados padronizados
pca = PCA(n_components=3)
pca.fit_transform(X)
print(pca.explained_variance_ratio_)
```

```
#PCA com dados padronizados
pca = PCA(n_components= 9)
principalComponents = pca.fit_transform(X_std)
print(pca.explained_variance_ratio_)
```

```
pca_Df = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 
                                                             'PC8', 'PC9'])
```

```
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('número de componentes')
plt.ylabel('variâcia cumulativa explicada')
plt.show()
```

```
# Criando uma tupla de autovalores e autovetores
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Ordenação dos autovalores e autovetores (decrescente)
eig_pairs.sort(reverse=True)

# Resultado dessa listagem:
print('Autovalores em ordem descrescente:')
for i in eig_pairs:
    print(i[0])
```

```
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Componente Principal 1',fontsize=20)
plt.ylabel('Componente Principal 2',fontsize=20)
plt.title("Análise de Componentes Principais - tipo de grau",fontsize=20)

targets = ['Bacharelado', 'Tecnológico', 'Licenciatura']
colors = ['r', 'g','b']

for target, color in zip(targets,colors):
    indicesToKeep = y == target
    plt.scatter(pca_Df.loc[indicesToKeep, 'PC1']
               , pca_Df.loc[indicesToKeep, 'PC2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

### Fase 5 - Redução das dimensões

```
model = PCA(n_components=9).fit(X_std)
X_pc = model.transform(X_std)

# número de componentes
n_pcs= model.components_.shape[0]

# pegando o índice das caracteríísticas mais importantes de cada componente (variável)
most_important = [np.abs(model.components_[i]).argmax() for i in range(n_pcs)]

initial_feature_names = X.columns
#pegando os nomes
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]

dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}
```
